```python
import re
import csv
import json
import time
import requests as rqs
import pandas as pd
from bs4 import BeautifulSoup as bs


title_list = []
type_list  = []
intro_list = []
date_list  = []



url_prefix = 'https://movies.yahoo.com.tw/movieinfo_main/'


start_t = time.time()

for i in range(1, 2):
    # generate url
    url = url_prefix + str(i)
    
    # send request
    response = rqs.get(url)
    response.encoding = 'utf-8'

    # extract web content 
    web_content = response.text

    
    soup = bs(web_content,'lxml')
    
    # movie title
    try:
        m_title = soup.find('h1').text
        title_list.append(m_title)
    except:
        continue;
    
    '''
    m_title = soup.find('h1').text
    title_list.append(m_title)
    '''
    
    # movie types
    m_type_content = soup.find('div', class_='level_name_box')
    each_type_content = m_type_content.find_all('a', class_='gabtn', attrs={'data-ga':re.compile('.*')})
    
    temp = []
    for element in each_type_content:
        #print(str(element['data-ga'].strip('][').split(',')[2].replace("'","")))
        temp.append(str(element['data-ga'].strip('][').split(',')[2]).replace("'",""))
    

    type_list.append(temp)
    
    # movie intro
    m_intro = soup.find('span', id='story').text.replace(" ", "").replace("\n", "")
    intro_list.append(m_intro)
    
    m_date_content = soup.find('div', class_='movie_intro_info_r')
    m_date = m_date_content.find('span').text.replace("上映日期：", "")
    date_list.append(m_date)

end_t = time.time()
print("time: ", end_t - start_t, "s")
movie_dict = {'title': title_list, 'type': type_list, 'intro': intro_list, 'date': date_list} 
df = pd.DataFrame(movie_dict)    


'''
df = pd.DataFrame(list(zip(title_list, type_list, date_list)),
                  columns =['title', 'type', 'date'])

'''

df.to_csv("../HW02/test_2.csv", index=False, encoding= 'utf_8_sig')
print("done")

#print(title_list)
#print(type_list)
#print(intro_list)
#print(date_list)
```
